```python
import ast
```

```python
ted["rating_list"]=ted.ratings.apply(lambda x:ast.literal_eval(x))
```

```python
def get_all_ratings(x):
    num=0
    for d in x:
        num=num+d["count"]
    return num
```

```python
ted["all_ratings"]=ted.rating_list.apply(lambda x: get_all_ratings(x))
```

```python
ted[["event","all_ratings"]]
```

```python
def get_funn_ratings(x):
    num=0
    for d in x:
        if d["name"]=="Funny":
            num=num+d["count"]
    return num
```

```python
ted["funny_ratings"]=ted.rating_list.apply(lambda x: get_funn_ratings(x))
```

```python
ted[["speaker_occupation","funny_rate"]].sort_values("funny_rate",ascending=False).head(10)
```

```python
ted["funny_rate"]=ted.funny_ratings/ted.all_ratings
```

```python
ted.groupby("speaker_occupation").funny_rate.agg(["mean"]).sort_values("mean").tail(10)
```
